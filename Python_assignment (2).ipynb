7. Describe the different types of loops in Python and their use cases with examples.

>>>Python has two main types of loops:
1️. for loop
2️. while loop

1️. For Loop used when you want to repeat a block a specific number of times.

    Example--
     for i in range(5):
        print(i)

2️. while Loop used when you don’t know how many times to loop, but want to loop until a condition becomes False.

     Example--
      i = 1
      while i <= 3:
       print(i)
       i += 1
